```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr = df.corr()
sns.heatmap(corr, cmap = "BuPu", annot = True);
```

```python
life_filter = df["healthy_life_expectancy"]
choice_filter = df["freedom_to_make_life_choices"]
corruption_filter = df["perceptions_of_corruption"]
social_filter = df["social_support"]
```

```python
plt.scatter(life_filter, choice_filter);
plt.title("Does freedom help you live longer")
plt.xlabel("Life Expectancy")
plt.ylabel("Freedowm to make choices")
plt.show()
#slight positive correlation between life expectancy and freedom of choices
```
